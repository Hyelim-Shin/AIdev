### 2. LSTM (Long Short-Term Memory)

* 입력 게이트: 장기 기억에 저장될 정보를 결정
* 삭제 게이트: 장기 기억에서 어떤 정보를 유지하거나 폐기할지 결정
* 출력 게이트: 현재 입력, 이전 단기 메모리 및 새로 계싼된 장기 메모리를 사용하여 다음 시간 단계에서 셀로 전달될 새로운 단기 메모리를 생성

**사용법**
```python
nn.LSTM(input_dim, hidden_size, batch_first=True)
```

### 3. GRU (Gated Recurrent Unit)

* 업데이트 게이트: 전달해야 하는 이전 정보의 양을 결정
* 리셋 게이트: 모델에서 무시해야 할 과거 정보의 양을 결정

**사용법**
```python
nn.GRU(input_dim, hidden_size, batch_first=True)
```
