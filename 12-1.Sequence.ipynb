## 순차 데이터 (Sequential Data/Sequence)

### 1. 순차 데이터 모델링

* 일반적으로 지도 학습을 위한 머신 러닝 알고리즘은 입력 데이터가 **독집 동일 분포** (Independent and Identically Distriuted, IID) 라고 가정
* 시퀀스를 다룰 때는 이러한 가정이 유효하지 않음
    - Ex) 주식 가격 예측, 날씨 예측

### 2. 순차 데이터 vs 시계열 데이터

* 시계열(time series) 데이터는 순차 데이터의 한 종류
    - 주식 가격, 녹화된 음성 등
* 모든 순차 데이터가 시간 차원을 가지는 것은 아님
    - 텍스트 데이터, DNA 시퀀스

### 3. 시퀀스 모델링의 종류

```python
Image(url='https://raw.githubusercontent.com/rickiepark/ml-with-pytorch/main/ch15/figures/15_02.png', width=500)
```

* 다대일(many-to-one)
    - 입력: 시퀀스, 출력: 고정크기의 텐서
    - Ex) 감성분석: 입력 - 텍스트, 출력 - 클래스 레이블 (좋음/싫음)
* 일대다(one-to-may)
    - 입력: 고정크기의 텐서, 출력: 시퀀스
    - Ex) 이미지 캡셔닝: 입력 - 이미지, 출력 - 이미지 요약 텍스트
* 다대다(many-to-many)
    - 입력: 시퀀스, 출력: 시퀀스
    - 동기적, 비동기적으로 나눌 수 있음
    - Ex) 품사 태깅(동기적): 입력 - 문장, 출력 - 각 단어의 품사
    - Ex) 번역(비동기적, 지연 존재): 입력 - 문장(A 언어), 출력 - 문장(B 언어)
